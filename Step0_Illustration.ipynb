```python
import flask
from flask import Flask, render_template_string, request, redirect, url_for
from threading import Thread

app = Flask(__name__)

# 学生数据
students = [
    {"id": 1, "name": "Alice", "gender": "女", "math": 85},
    {"id": 2, "name": "Bob", "gender": "男", "math": 78}
]

# 教师数据
teachers = [
    {"id": 1, "name": "Mr. Smith", "gender": "男", "subject": "Math"},
    {"id": 2, "name": "Ms. Johnson", "gender": "女", "subject": "English"}
]

# 首页菜单
@app.route("/")
def menu():
    template = """
    <h1>教务管理系统</h1>
    <p>请选择要查看的内容：</p>
    <a href="/students"><button style="padding:10px;margin:5px;">学生列表</button></a>
    <a href="/teachers"><button style="padding:10px;margin:5px;">教师列表</button></a>
    """
    return render_template_string(template)

# 学生列表
@app.route("/students")
def show_students():
    template = """
    <h1>学生列表</h1>
    <ul>
    {% for s in students %}
        <li>{{ s.id }} - {{ s.name }} ({{ s.gender }}) - {{ s.math }}分
        <a href="/delete_student/{{ s.id }}"><button>删除</button></a></li>
    {% endfor %}
    </ul>
    <h2>添加学生</h2>
    <form method="post" action="/add_student">
        姓名: <input type="text" name="name"><br>
        性别: 
        <select name="gender">
            <option value="男">男</option>
            <option value="女">女</option>
        </select><br>
        数学成绩: <input type="number" name="math"><br>
        <input type="submit" value="提交">
    </form>
    <a href="/"><button>返回首页</button></a>
    """
    return render_template_string(template, students=students)

# 添加学生
@app.route("/add_student", methods=["POST"])
def add_student():
    name = request.form["name"]
    gender = request.form["gender"]
    math = int(request.form["math"])
    new_id = (students[-1]["id"] + 1) if students else 1
    students.append({"id": new_id, "name": name, "gender": gender, "math": math})
    return redirect(url_for("show_students"))

# 删除学生
@app.route("/delete_student/<int:student_id>")
def delete_student(student_id):
    global students
    students = [s for s in students if s["id"] != student_id]
    return redirect(url_for("show_students"))

# 教师列表
@app.route("/teachers")
def show_teachers():
    template = """
    <h1>教师列表</h1>
    <ul>
    {% for t in teachers %}
        <li>{{ t.id }} - {{ t.name }} ({{ t.gender }}) - 教 {{ t.subject }}
        <a href="/delete_teacher/{{ t.id }}"><button>删除</button></a></li>
    {% endfor %}
    </ul>
    <h2>添加教师</h2>
    <form method="post" action="/add_teacher">
        姓名: <input type="text" name="name"><br>
        性别: 
        <select name="gender">
            <option value="男">男</option>
            <option value="女">女</option>
        </select><br>
        科目: <input type="text" name="subject"><br>
        <input type="submit" value="提交">
    </form>
    <a href="/"><button>返回首页</button></a>
    """
    return render_template_string(template, teachers=teachers)

# 添加教师
@app.route("/add_teacher", methods=["POST"])
def add_teacher():
    name = request.form["name"]
    gender = request.form["gender"]
    subject = request.form["subject"]
    new_id = (teachers[-1]["id"] + 1) if teachers else 1
    teachers.append({"id": new_id, "name": name, "gender": gender, "subject": subject})
    return redirect(url_for("show_teachers"))

# 删除教师
@app.route("/delete_teacher/<int:teacher_id>")
def delete_teacher(teacher_id):
    global teachers
    teachers = [t for t in teachers if t["id"] != teacher_id]
    return redirect(url_for("show_teachers"))

# 启动服务（5001 避免冲突）
Thread(target=lambda: app.run(port=5006, debug=False, use_reloader=False)).start()
```
